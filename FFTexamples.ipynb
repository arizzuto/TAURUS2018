```python
#Feel free to play with inputs of t, frequncy, 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#example 1: Simple sine wave FFT
#First make a sine wave:
t=np.arange(0,1,.005)
f=5.
omega = f*(2.*np.pi)
omega
A=20.
phi=45*np.pi/180.
y=A*np.sin(omega*t+phi)
plt.plot(t,y)
plt.show()
```

```python
#Set up frequency space and take FFT
Fs=1/.005 #sample frequency
n=len(y) #length of vector
YY=np.fft.fft(y)/n
k=np.arange(n)
TT=n/Fs #Total time of signal
frq=k/TT  #convert time to frequency space
frq=frq[range(n/2)] #we only need half of the vector
YY=YY[range(n/2)]
plt.plot(frq,np.abs(YY)) #We plot abs value of YY because it contain complex #s
plt.show()
#Notice that the frequency lines up with your input frequency. Yay!
```

```python
#example 2: Mixed frequency signal
t=np.arange(0,1,.01)
f1=20.
f2=2.
omega1 = f1*(2.*np.pi)
omega2 = f2*(2.*np.pi)
A1=1
A2=5.5
#phi1=45*np.pi/180.
phi1=0.
phi2=0.
```
